Вычислить сумму трех произвольных действительных чисел.

```python
from keras.models import Sequential
from keras.layers import Dense
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

```python
train_input = np.random.randint(500, size=(200, 3))
train_output = np.array([sum(i) for i in train_input])
test_input = np.random.randint(500, size=(50, 3))
test_output =  np.array([sum(i) for i in test_input])
```

```python
model = Sequential()
model.add(Dense(units=1, input_shape=(3,), activation='relu'))
model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(learning_rate=0.1))
fit_model = model.fit(train_input, train_output, epochs=500, verbose=False)
predictions = model.predict(test_input)
```

```python
print(predictions)
print(test_output)
```

```python
print(model.get_weights())
```

```python
print(fit_model.history['loss'])
```
